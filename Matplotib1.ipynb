# MatplotLib.pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simple ploting
```

```python
plt.plot([1,2,3,4])
#the above provided list is Y Axis
#X-Axis is the indices = [0,1,2,3]
plt.ylabel('numbers')
plt.xlabel('indices')
plt.title('my plot')
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
plt.plot(x,y,'ro')
plt.grid()  #shows grid in graph
plt.show()
```

```python
t1=np.arange(0,5,0.2)
plt.plot(t1, t1**2, 'b--', label='^2', linewidth=3.0)
plt.plot(t1, t1**3, 'r^', label='^3', linewidth=5.0)
plt.grid()
plt.legend()
plt.show()
```

```python
x1=[1,2,3,4]
x2=[1,4,9,16]
x3=[2,4,6,8]

lines=plt.plot(x1,x2,x1,x3)
#first 2 args are lines[0], and second 2 are lines[1]
#we can the change the properties of these lines separately

# using keyword args
plt.setp(lines[0], color='r', linewidth=3.0)

# MATLAB style String value pairs
plt.setp(lines[1], 'color','g', 'linewidth', 5.0)
plt.grid()
plt.show()
```

# working with multiple figures and axes

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi**t)

t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.02)

plt.figure(1)  #it creates 1 figure

plt.subplot(211)
plt.plot(t1,f(t1), 'b-')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2), 'r--')

plt.show()
```
